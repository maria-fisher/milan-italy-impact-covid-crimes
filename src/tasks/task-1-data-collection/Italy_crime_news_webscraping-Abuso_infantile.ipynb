```python
from bs4 import BeautifulSoup as bs
import requests
import re
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from datetime import date
today = date.today()
d = today.strftime("%m-%d-%y")

```

```python
news_p1 = requests.get("https://gazzettadelsud.it/ricerca/?f=Abuso+infantile")
news_p1
```

```python
news_p1 = bs(news_p1.content)
news_p1
```

```python
news_p1.find_all(class_="titolo") 
```

```python
titles1 = [i.text for i in news_p1.find_all(class_='titolo')]
titles1
```

```python
news_p1.find_all(class_="occhiello") 
```

```python
provincia = [i.text for i in news_p1.find_all(class_='nome_prov')]
provincia
```

```python
date1 = [i.text for i in news_p1.find_all(itemprop="dateCreated datePublished")]
date1
```

```python
date1 = date1
date_list1 = []
for element in date1:

    date_list1.append(element.strip())


print(date_list1)
```
